Suppose we clumsily decide to perform an **inner join merge** of the two dataframes on just the `first_name` column. Do this below and show the resulting dataframe.

### 1.3. Explaining the Result

How would you explain to a `STAT107 student` (who has never performed an **inner join merge** before) why the resulting merged dataframe in 1.2 resulted in 11 rows?

An inner join merge matches each primary key from the left dataframe with each primary key in the right dataframe. If the key matches then the merged row is included in the dataframe. If the key does not match, it is not included in the dataframe. As we only used first_name as the primary key, every first name from the left dataframe name matches with the same first name in the right dataframe, regardless of whether it is the correct person. This causes 11 rows.

### 1.4. Outer Join Merge

Suppose we also clumsily decide to perform an **outer join merge** of the two dataframes on just the `first_name` column. Do this below and show the resulting dataframe.

```python
df_secondary = pd.merge(df_occupation, df_generation, how='outer', left_on=['first_name'], right_on=['first_name'])
df_secondary
```

### 1.5. Explaining the Result

How would you explain to a `STAT107 student` (who has never performed an **outer join merge** before) why the resulting merged dataframe in 1.4 resulted in 13 rows?

An outer join merge matches each primary key from the left dataframe with each primary key in the right dataframe. If the key matches then the merged is included in the dataframe. If the key does not match, then both the unmatched rows are added into the final dataframe merged with NaN values. As we only used first_name as the primary key, every first name from the left dataframe name matches with the same first name in the right dataframe, regardless of whether it is the correct person. This causes 13 rows.

## 2. Linear Regression with Categorical Explanatory Variables

Suppose we like to build a linear regression model that predicts `number_of_Instagram_followers` given:
* `number_of_posts`
* `account_type`: personal, creator, business
* `have_profile_picture`: yes, no

Explain to a STAT107 student why we would seek to find the optimal intercept and slopes for THIS equation below...

\begin{align*}
\hat{\text{number\_of\_Instagram\_followers}} &= \hat{\beta}_0 \\
&+ \hat{\beta}_1 \times number\_of\_posts \\
&+ \hat{\beta}_2 \times account\_type_{personal} \\
&+ \hat{\beta}_3 \times account\_type_{creator} \\
&+ \hat{\beta}_4 \times have\_profile\_picture_{yes}
\end{align*}


... as opposed to THIS equation below...

\begin{align*}
\hat{number\_of\_Instagram\_followers} &= \hat{\beta}_0 \\
&+ \hat{\beta}_1 \times number\_of\_posts \\
&+ \hat{\beta}_2 \times account\_type_{personal} \\
&+ \hat{\beta}_3 \times account\_type_{creator} \\
&+ \hat{\beta}_4 \times account\_type_{business} \\
&+ \hat{\beta}_5 \times have\_profile\_picture_{yes} \\
&+ \hat{\beta}_6 \times have\_profile\_picture_{no}
\end{align*}


The first equation is better because it helps avoid multicollinearity and redundancy. By omitting one category for each categorical variable, we remove a category which is highly correlation with the rest of the categories helping us avoid multicollinearity. The model also becomes simpler to interpret as each coefficient represents the number of followers compared to the baseline category. The baseline category  is the category we dropped, in this case business and no.

1. We will have misleading slope interpretations. The interpretation of some slopes becomes misleading and does not represent the actual relationship between the exlanatory and response variable.
2. Overfitting. By including both collinear explanatory pairs, it may increase the training dataset R^2 but decrease the test dataset R^2.
